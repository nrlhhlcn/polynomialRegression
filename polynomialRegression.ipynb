```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Position_Salaries.csv")
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4) #bu her zaman 2 degildir veri setine göre degişebilir
X_poly=poly_reg.fit_transform(X)
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_poly,Y)
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,lin_reg.predict(X),color="blue")
plt.title("Doğru mu Blof mü(liner )")
plt.xlabel("Pozisyon Seviyeleri")
plt.ylabel("Maaş")
plt.show()
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color="blue")
plt.title("Doğru mu Blöf Mü (Polinom)")
plt.xlabel("Pozisyon Seviyeleri")
plt.ylabel("Maaş")
plt.show()
```

```python
lin_reg.predict([[6.5]]) #dogrusal regrsyon ile böyle bir sonuç cıkıyor
```
